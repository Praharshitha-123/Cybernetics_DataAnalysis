```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Ubysoft Cybernetics Dataset.csv')
df[['Platform', 'Developers','Name of the Game']] = df[['Platform', 'Developers','Name of the Game']].astype(str)
df['Release Date']= df['Release Date'].apply(pd.to_datetime)
print(df.dtypes)
```

```python
#Breaking the Platform data into individual entries.
def splitDataFrameList(df,target_column,separator):
    
    row_accumulator = []

    def splitListToRows(row, separator):
        split_row = row[target_column].split(separator)
        for s in split_row:
            new_row = row.to_dict()
            new_row[target_column] = s
            row_accumulator.append(new_row)

    df.apply(splitListToRows, axis=1, args = (separator, ))
    new_df = pd.DataFrame(row_accumulator)
    return new_df
df1 = splitDataFrameList(df,'Platform','||')
```

```python
#df2 contains plaform and total number of games on each platform
df2=df1.groupby('Platform').count()
df2= df2.drop(['Release Date','Website','User Rating','Metacritic Rating','ESRB Rating','Playtime','Achievements Count','Ratings Count','Suggestions Count','Game Series Count','Reviews Count','Developers','Genres','Not Played','Owned','Completed','Playing'],axis=1)
df2.rename(columns={'Name of the Game':'Total_no_of_games'},inplace=True)
df2.sort_values(by='Platform',ascending=True)
print(df2)
```

```python
#df_r contains year wise games released on each platform
df_n=df1[['Platform','Name of the Game','Release Date']]
import datetime 
df_n['release_year']=pd.DatetimeIndex(df_n['Release Date']).year
df_r=pd.pivot_table(df_n,values='Name of the Game',index='Platform',columns='release_year',aggfunc=len)
df_r=df_r.round(decimals=0)
df_r
```

```python
# merging df_r and df_2 to get dfq1, and also findind what percentage of total games are released on each platform.
dfq1=pd.merge(df2,df_r,on='Platform',how='outer')
dfq1['percentage_of_games']=round((dfq1['Total_no_of_games']/(dfq1['Total_no_of_games'].sum()))*100,3)
dfq1=dfq1.sort_values(by='Total_no_of_games',ascending=False)
dfq1 = dfq1.fillna(0)
dfq1
#dfq1.to_csv('ques1_table.csv')
```

```python
#list of platforms that have seen a YoY growth in number of games released on them.
for x in range(dfq1.shape[0]):
    k=1
    
    for i in range(1,12):
        if dfq1.iat[x,i]<dfq1.iat[x,i+1]:
            continue
        else:
            k=0   
    if k==1:
        print(dfq1.index[x])    
```

```python
# list of platforms that became redundant in the 2010s decade(less than 10 games released/year)
for x in range(dfq1.shape[0]):
    k=1
    
    for i in range(4,13):
        if dfq1.iat[x,i]<10:
            continue
        else:
            k=0   
    if k==1:
        print(dfq1.index[x]) 
```

```python
#Platform with highest growth over the years 2007-2018 is Web. Xbox, Commodore, Atari and Playstation5 have inf % since they had 0 games in 2007 
dfq1['growth_over_time']= round(((dfq1[2018.0]-dfq1[2007.0])/dfq1[2007.0])*100,1)
dfq1['growth_over_time'].sort_values(ascending=False)
```
