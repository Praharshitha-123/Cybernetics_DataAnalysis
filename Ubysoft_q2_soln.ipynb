```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Ubysoft Cybernetics Dataset.csv')
df[['Genres', 'Developers','Name of the Game']] = df[['Genres', 'Developers','Name of the Game']].astype(str)
df['Release Date']= df['Release Date'].apply(pd.to_datetime)
print(df.dtypes)
```

```python
def splitDataFrameList(df,target_column,separator):
    
    row_accumulator = []

    def splitListToRows(row, separator):
        split_row = row[target_column].split(separator)
        for s in split_row:
            new_row = row.to_dict()
            new_row[target_column] = s
            row_accumulator.append(new_row)

    df.apply(splitListToRows, axis=1, args = (separator, ))
    new_df = pd.DataFrame(row_accumulator)
    return new_df
df1 = splitDataFrameList(df,'Genres','||')
df1['ESRB Rating']=df1['ESRB Rating'].fillna('Not Rated')
df1['User Rating']=df1['User Rating'].replace(0,np.nan)
df1['Playtime']=df1['Playtime'].replace(0,np.nan)
```

```python
df_a = pd.pivot_table(df1,index='Genres',values='Name of the Game',aggfunc=len)
```

```python
df_b = round(pd.pivot_table(df1,index='Genres',values=['User Rating','Metacritic Rating','Playtime']),2)
```

```python
df_c = pd.pivot_table(df1,index='Genres',values='Name of the Game',columns='ESRB Rating',aggfunc=len)
df_c=df_c.fillna(0)
```

```python
df_d= pd.merge(df_a,df_b,on='Genres',how='outer')
```

```python
dfq2=pd.merge(df_d,df_c,on='Genres',how='outer')
dfq2
```

```python
#the top 4 genres according to User rating 
dfq2.sort_values(by='User Rating',ascending=False)[0:4]
```

```python
#the top 4 genres according to Metacritic rating 
dfq2.sort_values(by='Metacritic Rating',ascending=False)[0:4]
```

```python
#mean game playtime across all genres
round(dfq2['Playtime'].mean(),2)
```
