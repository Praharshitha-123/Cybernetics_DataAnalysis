```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Ubysoft Cybernetics Dataset.csv')
df[['Genres', 'Developers','Name of the Game']] = df[['Genres', 'Developers','Name of the Game']].astype(str)
df['Release Date']= df['Release Date'].apply(pd.to_datetime)
```

```python
def splitDataFrameList(df,target_column,separator):
    
    row_accumulator = []

    def splitListToRows(row, separator):
        split_row = row[target_column].split(separator)
        for s in split_row:
            new_row = row.to_dict()
            new_row[target_column] = s
            row_accumulator.append(new_row)

    df.apply(splitListToRows, axis=1, args = (separator, ))
    new_df = pd.DataFrame(row_accumulator)
    return new_df
df1 = splitDataFrameList(df,'Genres','||')
```

```python
df1 = splitDataFrameList(df1,'Developers','||')
df1['Completed']=df1['Completed'].replace(0,np.nan)
df1['Owned']=df1['Owned'].replace(0,np.nan)
df1['ESRB Rating']=df1['ESRB Rating'].fillna('Not Rated')
df1['User Rating']=df1['User Rating'].replace(0,np.nan)
df1
```

```python
#Developers with most number of games released.
df_a = pd.pivot_table(df1,index='Developers',values='Name of the Game',aggfunc=len)
df_a=df_a.sort_values(by='Name of the Game',ascending=False)[1:11]
df_a
```

```python
#developers and the number of games released by them across genres.
df_b = pd.pivot_table(df1,index='Developers',values='Name of the Game',columns='Genres',aggfunc=len)
df_b['Total_games']=df_b.sum(axis=1)
df_b=df_b.sort_values(by='Total_games',ascending=False).fillna(0)
df_b
```
